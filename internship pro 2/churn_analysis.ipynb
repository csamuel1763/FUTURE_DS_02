```python
import pandas as pd

df = pd.read_csv("churn.csv")

print(df.shape)
print(df.head())
total_customers = len(df)
print(total_customers)
churn_customers = len(
    df[df["Churn"]=="Yes"]
)

print(churn_customers)
churn_rate = (
    churn_customers /
    total_customers
) * 100

print(round(churn_rate,2))
def tenure_group(x):
    if x <= 12:
        return "0-1 Year"
    elif x <= 24:
        return "1-2 Years"
    elif x <= 36:
        return "2-3 Years"
    elif x <= 48:
        return "3-4 Years"
    elif x <= 60:
        return "4-5 Years"
    else:
        return "5+ Years"

df["TenureGroup"] = df["tenure"].apply(
    tenure_group
)
df.to_csv(
    "cleaned_churn.csv",
    index=False
)
df.info()
df.isnull().sum()
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```
